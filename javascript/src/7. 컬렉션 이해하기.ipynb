**ğŸ”· 1) map - ë°°ì—´ì„ 'ë‹¤ë¥¸ í˜•íƒœ'ë¡œ ë³€í™˜í•  ë•Œ**

````typescript
const products = [
  { id: 101, name: 'Laptop', price: 1200 },
  { id: 102, name: 'Mouse', price: 25 },
];

// ì˜ˆì‹œ 1: ê° ìƒí’ˆì˜ ì„¸ê¸ˆ í¬í•¨ ê°€ê²©(finalPrice)ì„ ê³„ì‚°í•œ ìƒˆë¡œìš´ ë°°ì—´ ìƒì„±
const productsWithTax = products.map(product => ({
  ...product,                  // ê¸°ì¡´ ì†ì„± ë³µì‚¬ (id, name, price ê·¸ëŒ€ë¡œ ìœ ì§€)
  finalPrice: Math.round(product.price * 1.1),  // 10% ì„¸ê¸ˆ í¬í•¨í•œ ê°€ê²© ì¶”ê°€
}));

console.log(productsWithTax);

// ê²°ê³¼:
// [
//   { id: 101, name: 'Laptop', price: 1200, finalPrice: 1320 },
//   { id: 102, name: 'Mouse', price: 25, finalPrice: 28 }
// ]
````

````typescript
const userList = [
  { name: 'Alice', status: 'active', role: 'admin' },
  { name: 'Bob', status: 'inactive', role: 'user' },
  { name: 'Charlie', status: 'active', role: 'user' },
];

// ì˜ˆì‹œ 2: í™œì„±(active) ìƒíƒœ & ì—­í• (role)ì´ userì¸ ì‚¬ìš©ìë§Œ ì¶”ì¶œ
const activeUsers = userList.filter(user =>
  user.status === 'active' && user.role === 'user'
);

console.log(activeUsers);

// ê²°ê³¼:
// [ { name: 'Charlie', status: 'active', role: 'user' } ]
````

**ğŸ”· 3) reduce - ì—¬ëŸ¬ ê°’ì„ í•˜ë‚˜ì˜ ê²°ê³¼ë¡œ 'ëˆ„ì ' ë˜ëŠ” 'ìš”ì•½'í•  ë•Œ**  
reduceëŠ” ë°°ì—´ ì „ì²´ë¥¼ ìˆœíšŒí•˜ë©° ê°’ì„ ëˆ„ì (accumulate)í•˜ì—¬ ë‹¨ í•˜ë‚˜ì˜ ê²°ê³¼ë¥¼ ë§Œë“¤ì–´ ëƒ…ë‹ˆë‹¤.

**âœ”ï¸ reduce ì˜ˆì œ A: ìˆ«ì ëˆ„ì **

````typescript
const items = [
  { item: 'A', count: 3 },
  { item: 'B', count: 5 },
  { item: 'C', count: 1 },
];

// ì˜ˆì‹œ 3A: ëª¨ë“  count ê°’ì„ ë”í•˜ì—¬ ì´í•© êµ¬í•˜ê¸°
const totalCount = items.reduce((accumulator, current) => {
  return accumulator + current.count; // í˜„ì¬ í•­ëª©ì˜ countë¥¼ ëˆ„ì 
}, 0); // ì´ˆê¸°ê°’(initialValue): 0ë¶€í„° ëˆ„ì  ì‹œì‘

console.log(totalCount);
// ê²°ê³¼: 9
````

**âœ”ï¸ reduce ì˜ˆì œ B: ë°°ì—´ì„ 'ê°ì²´ í˜•íƒœ'ë¡œ ë³€í™˜í•˜ê¸°**

````typescript
// ì˜ˆì‹œ 3B: item ê°’ì„ keyë¡œ í•˜ê³  countë¥¼ valueë¡œ í•˜ëŠ” ê°ì²´ë¡œ ë³€í™˜
const itemsMap = items.reduce((acc, current) => {
  acc[current.item] = current.count;  // ê°ì²´ì— ë™ì ìœ¼ë¡œ key ì¶”ê°€
  return acc;
}, {}); // ì´ˆê¸°ê°’: ë¹ˆ ê°ì²´ë¡œ ì‹œì‘

console.log(itemsMap);
// ê²°ê³¼:
// { A: 3, B: 5, C: 1 }
````

1\. ì˜ˆì‹œ: Setì˜ ë‹¤ì–‘í•œ ë©”ì„œë“œ í…ŒìŠ¤íŠ¸ ì½”ë“œ


````typescript
// -------------------------------------------
// 1. Map ìƒì„± & ê¸°ë³¸ set/get
// -------------------------------------------

// í‚¤ë¡œ ì‚¬ìš©í•  ë‹¤ì–‘í•œ íƒ€ì…
const objKey = { type: 'objectKey' };
const arrKey = ['arrayKey'];
const fnKey = function fnKey() {};

const map = new Map();

// set(): í‚¤-ê°’ ì €ì¥
map.set(objKey, { message: 'Object key ì‚¬ìš© ì˜ˆì‹œ' });
map.set(arrKey, { message: 'Array key ì‚¬ìš© ì˜ˆì‹œ' });
map.set(fnKey, { message: 'Function key ì‚¬ìš© ì˜ˆì‹œ' });

// get(): í‚¤ë¡œ ê°’ ì¡°íšŒ
console.log('--- get() í…ŒìŠ¤íŠ¸ ---');
console.log(map.get(objKey)); // { message: 'Object key ì‚¬ìš© ì˜ˆì‹œ' }
console.log(map.get(arrKey)); // { message: 'Array key ì‚¬ìš© ì˜ˆì‹œ' }
console.log(map.get(fnKey));  // { message: 'Function key ì‚¬ìš© ì˜ˆì‹œ' }


// -------------------------------------------
// 2. has(): í‚¤ ì¡´ì¬ ì—¬ë¶€ ë¹ ë¥´ê²Œ í™•ì¸
// -------------------------------------------
console.log('\n--- has() í…ŒìŠ¤íŠ¸ ---');
console.log(map.has(arrKey)); // true
console.log(map.has('notExist')); // false


// -------------------------------------------
// 3. size: Map ìš”ì†Œ ê°œìˆ˜ â†’ O(1)ë¡œ ì„±ëŠ¥ ë§¤ìš° ì¢‹ìŒ
// -------------------------------------------
console.log('\n--- size í…ŒìŠ¤íŠ¸ ---');
console.log(`Map í¬ê¸°: ${map.size}ê°œ`);


// -------------------------------------------
// 4. delete(): íŠ¹ì • í‚¤ ì‚­ì œ
// -------------------------------------------
console.log('\n--- delete() í…ŒìŠ¤íŠ¸ ---');
map.delete(fnKey);
console.log(map.has(fnKey)); // false â†’ ì‚­ì œ í™•ì¸
console.log(`ì‚­ì œ í›„ Map í¬ê¸°: ${map.size}ê°œ`);


// -------------------------------------------
// 5. keys(), values(), entries(): ìˆœíšŒ ê´€ë ¨ ë©”ì„œë“œ
// -------------------------------------------
console.log('\n--- keys(), values(), entries() ìˆœíšŒ ---');

console.log('keys():');
for (const key of map.keys()) {
  console.log('  key =', key);
}

console.log('values():');
for (const value of map.values()) {
  console.log('  value =', value);
}

console.log('entries():');
for (const [key, value] of map.entries()) {
  console.log(`  key = ${JSON.stringify(key)}, value = ${JSON.stringify(value)}`);
}


// -------------------------------------------
// 6. forEach(): Mapì—ì„œë„ forEach ì‚¬ìš© ê°€ëŠ¥
// -------------------------------------------
console.log('\n--- forEach() ìˆœíšŒ ---');
map.forEach((value, key) => {
  console.log(`key: ${JSON.stringify(key)}, value: ${JSON.stringify(value)}`);
});


// -------------------------------------------
// 7. clear(): ì „ì²´ ìš”ì†Œ ì‚­ì œ
// -------------------------------------------
// map.clear();
// console.log(`clear() í›„ size = ${map.size}`);
````

2\. ì˜ˆì‹œ: Mapì„ í™œìš©í•œ ìºì‹œ(Cache) íŒ¨í„´

````typescript
const cache = new Map();

function heavyCompute(num) {
  console.log('ğŸ’¡ ë¬´ê±°ìš´ ì‘ì—… ì‹¤í–‰');
  return num * num;
}

function getCachedValue(num) {
  if (cache.has(num)) {
    return cache.get(num); // ìºì‹œëœ ê°’ ë°˜í™˜
  }
  const result = heavyCompute(num);
  cache.set(num, result);
  return result;
}

console.log(getCachedValue(10)); // ë¬´ê±°ìš´ ì‘ì—… ì‹¤í–‰ â†’ 100
console.log(getCachedValue(10)); // ìºì‹œ HIT â†’ ë¹ ë¥´ê²Œ 100
````

### **4\. Map & Set í™•ì¥ ê¸°ëŠ¥ (ES2024)**

íŠ¹íˆ Object.groupBy / Map.groupByëŠ” ê¸°ì¡´ì— reduceë¥¼ ì‚¬ìš©í•´ ì–´ë µê²Œ ì‘ì„±í•˜ë˜ "ê·¸ë£¹í™”(grouping)" ì‘ì—…ì„ ë§¤ìš° ê°„ê²°í•˜ê²Œ ë§Œë“¤ê³ , Setì˜ ì§‘í•© ì—°ì‚°(union, intersection, difference)ì€ ìë£Œ ë¶„ì„Â·ê¶Œí•œ ì²˜ë¦¬Â·í•„í„°ë§ ë“± ë‹¤ì–‘í•œ ì‹¤ë¬´ ìƒí™©ì—ì„œ ìœ ìš©í•©ë‹ˆë‹¤.

ğŸ”· **1) Object.groupBy / Map.groupBy (ES2024)**  
ë°°ì—´ì˜ ìš”ì†Œë¥¼ ì–´ë–¤ ê¸°ì¤€(í‚¤)ìœ¼ë¡œ ê·¸ë£¹í™”í•˜ì—¬ ë¬¶ì–´ì£¼ëŠ” ê¸°ëŠ¥ì…ë‹ˆë‹¤.

````typescript
//ì˜ˆì œ: status(ìƒíƒœ)ë³„ ê·¸ë£¹í™”
const orders = [
  { id: 1, status: 'pending', region: 'Seoul' },
  { id: 2, status: 'shipped', region: 'Busan' },
  { id: 3, status: 'pending', region: 'Seoul' },
];

// Object.groupBy â†’ ê²°ê³¼ëŠ” ì¼ë°˜ ê°ì²´
//order => order.status
//function (order) {
//  return order.status;
//}
//ì›í•˜ëŠ” ëª¨ë“  ì¡°ê±´ / ë¡œì§ / ê³„ì‚°ì‹ì„ ë„£ì„ ìˆ˜ ìˆìŠµë‹ˆë‹¤.

const ordersByStatus = Object.groupBy(orders, order => order.status);
console.log(ordersByStatus);
/*
ê²°ê³¼:
{
  pending: [
    { id: 1, status: 'pending', region: 'Seoul' },
    { id: 3, status: 'pending', region: 'Seoul' }
  ],
  shipped: [
    { id: 2, status: 'shipped', region: 'Busan' }
  ]
}
*/
console.log(ordersByStatus.pending.length); // 2
````

**ğŸ”· 2) Set ì—°ì‚° ë©”ì„œë“œ (ES2024)**  
Setì€ ì›ë˜ â€œê³ ìœ í•œ ê°’ë“¤ì˜ ì§‘í•©â€ì„ ì €ì¥í•˜ëŠ” ìë£Œêµ¬ì¡°ì…ë‹ˆë‹¤.

| ë©”ì„œë“œ | ì˜ë¯¸ | ì„¤ëª… |  
| --- | --- | --- |  
| union(other) | í•©ì§‘í•© A âˆª B | ë‘ Setì˜ ëª¨ë“  ìš”ì†Œë¥¼ í•©ì³ ìƒˆë¡œìš´ Set ë°˜í™˜ |

````typescript
const userPermissions = new Set(['read', 'write', 'admin']);
const defaultPermissions = new Set(['read', 'guest']);

// intersection: ê³µí†µ ê¶Œí•œ (A âˆ© B)
const commonRights = userPermissions.intersection(defaultPermissions);
/*
Set { 'read' }
*/
console.log([...commonRights]);  // ['read']

// difference: ì‚¬ìš©ìì—ê²Œë§Œ ìˆëŠ” ê¶Œí•œ (A \ B)
const exclusiveRights = userPermissions.difference(defaultPermissions);
/*
Set { 'write', 'admin' }
*/
console.log([...exclusiveRights]); // ['write', 'admin']

// union: ì „ì²´ ê¶Œí•œ ëª©ë¡ (A âˆª B)
const allRights = userPermissions.union(defaultPermissions);
/*
Set { 'read', 'write', 'admin', 'guest' }
*/
console.log([...allRights]); // ['read', 'write', 'admin', 'guest']
````
